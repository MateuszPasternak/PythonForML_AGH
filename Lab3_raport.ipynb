Task 11

```python
from sklearn.model_selection import train_test_split

q_low = data['BMI'].quantile(0.01) # get 1st percentile
q_hi = data['BMI'].quantile(0.99) # get 99th percentile
data_filtered = data[(data['BMI'] < q_hi) & (data['BMI'] > q_low)]

X = data_filtered.drop(['Heart Rate','Dream Weight'], axis = 1)
y = data_filtered['Heart Rate']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=1)

X_train.head()
```
